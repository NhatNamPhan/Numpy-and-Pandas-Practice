```python
import pandas as pd
import numpy as np
```

```python
data = {'Tên':['Nam','Huy','Hoàng','Anh','Thịnh'],
        'Tuổi':[21,22,23,24,25],
        'Điểm':[10,5,9,7,8]}
df = pd.DataFrame(data)
df
```

```python
#Lọc các hàng có điểm > 8.
df.loc[df['Điểm']>8]
```

```python
#Thêm cột 'trạng thái' với giá trị 'Pass' nếu điểm >= 5, ngược lại là 'Fail'.
df['Trạng thái'] = ['Pass' if x >= 5 else 'Fail' for x in df['Điểm']]
df
```

```python
#Tạo một DataFrame có ít nhất 2 giá trị NaN và thay thế chúng bằng giá trị trung bình của cột.
data = {'col1':[1,2,np.nan,4],
        'col2':[2,np.nan,5,6],
        'col3':[np.nan,4,2,np.nan]}
df = pd.DataFrame(data)
df
```

```python
df_filled = df.fillna(df.mean(numeric_only=True))
df_filled
```
