```python
import pandas as pd
import numpy as np
```

```python
#Tạo hai DataFrame với một cột chung và thực hiện merge theo các kiểu: inner, left, right, outer.
data_1 = {'Name': ['Nam','An','Khoa','Duy','Dung','Ly'],
          'Age': [20, 23, 35, 10, 25, 19],
          'City': ['Da Nang','Ha Noi','New York','Ho Chi Minh','London','Paris']}
data_2 = {'Name': ['Nam','An','Khoa','Duy','Dung','Huy'],
          'Profession': ['Date Engineer','Accountant','Teacher','Student','Project Manager','Doctor'],
          'Salary': [100000, 500000, 1000000, 0, 500000, 100000]}
df_1 = pd.DataFrame(data_1)
df_2 = pd.DataFrame(data_2)
```

```python
merged_inner = pd.merge(df_1, df_2, on='Name', how='inner')
merged_inner
```

```python
merged_left = pd.merge(df_1, df_2, on='Name', how='left')
merged_left
```

```python
merged_right = pd.merge(df_1, df_2, on='Name', how='right')
merged_right
```

```python
merged_outer = pd.merge(df_1, df_2, on='Name', how='outer')
merged_outer
```

```python
#Nối hai DataFrame theo hàng và cột bằng concat.
combine_row = pd.concat([df_1,df_2], axis=0)
combine_row
```

```python
combine_column = pd.concat([df_1['City'],df_2['Profession']], axis=1)
combine_column 
```

```python
#Tạo một DataFrame với dữ liệu trùng lặp và loại bỏ chúng bằng drop_duplicates.
data_3 = {'Name':['Nam','Huy','Nam','Minh','Trang','An','Huy','Minh'],
          'Birthday': [2004,2005,2004,2008,1990,1992,2005,2008],
          'Height': [175,180,175,165,162,172,180,165]}
df_3 = pd.DataFrame(data_3)
df_3
```

```python
df_cleaned = df_3.drop_duplicates(subset=['Name','Birthday','Height'])
df_cleaned
```

```python
#Kết hợp 3 DataFrame có cột chung và kiểm tra kết quả.
merged = pd.merge(df_1,df_2,on='Name', how='outer')
final_merged = pd.merge(df_cleaned,merged, on='Name',how='outer')
final_merged
```
