```python
import pandas as pd
import numpy as np
```

```python
#Tạo một DataFrame với cột ngày tháng (10 ngày liên tiếp) và cột giá trị ngẫu nhiên.
dates = pd.date_range('2025-01-01', periods=10, freq='8D')
value = np.random.randint(1, 100, size=10)
df = pd.DataFrame({'date': dates, 'value': value})
df
```

```python
#Lấy ra ngày và tháng từ cột ngày tháng.
df['Day'] = [d.day_name() for d in dates]
df['Month'] = [m.month_name() for m in dates]
df
```

```python
#Tính tổng giá trị theo mỗi 8 ngày (resampling).
df.set_index('date').resample('8D').sum()
```

```python
#Tính trung bình trượt (rolling mean) với cửa sổ 3 ngày.
rolling_mean = df['value'].rolling(window=3).mean()
rolling_mean
```
